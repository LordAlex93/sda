-   Procedemos a cribar los números tomando el primero y tachando sus
    múltiplos, luego el siguiente que no hayamos tachado y así hasta
    concluir la lista.

``` python
def cribado_eratostenes(n):
    """
    Implementación del Cribado de Eratóstenes
    """
    total=0
    # Creamos una lista con todos los números del 2 al n
    numeros = list(range(2, n+1))
    # Iniciamos el cribado
    for i in numeros:
        # Si el número ya ha sido marcado como no primo, saltamos a la siguiente iteración
        if i is None:
            continue
        # Si el número es primo, marcamos todos sus múltiplos como no primos
        total+=1
        for j in range(i*2, n+1, i):
            numeros[j-2] = None
    # Devolvemos una lista con los números primos encontrados
    return total,[num for num in numeros if num is not None]
```

``` python
# Encontramos los primeros números primos
total,primos = cribado_eratostenes(20);
# Imprimimos los resultados
print(f"Total de números encontrados {total}, números primos {primos}")
```

> **Atención**
>
> Pero este no es el método óptimo ya que realizamos demasiadas
> iteraciones, hay otro mejor todavía:

``` python
# Criba de Eratóstenes optimizada para generar primos hasta n
def criba_eratostenes(n):
    if n < 2: return []
    primos = [True] * (n + 1) # Inicializamos una lista de tamaño n+1. 
    primos[0] = primos[1] = False
    for p in range(2, int(n**0.5) + 1):
        if primos[p]:
            for i in range(p * p, n + 1, p):
                primos[i] = False
    return [i for i, primo in enumerate(primos) if primo]
```

``` python
# Ejemplo de uso
print(criba_eratostenes(100)) # Genera primos hasta 100
```


-   Vamos a usar el algoritmo menos operativo de factorización, no
    obstante, es la definición propiamente:

``` python
def fact1(n):
    factores = []
    for i in range(1,n+1):
        if n%i == 0:
            factores.append(i)
    return factores
```

Por ello, nos conviene utilizar otro algoritmo de factorización mejor.

``` python
def factorizar(n):
    factores = []

    for divisor in range(2, int(n**0.5) + 1):
        while n % divisor == 0:
            factores.append(divisor)
            n //= divisor
    
    if n > 1:
        factores.append(n)
    return factores
```

``` python
factorizar(765)
```

Aplicamos ahora un ejemplo con estilo para factorizar un número:

``` python
numero = 765
factores = factorizar(numero)
print(f"{numero} = {'*'.join(map(str, factores))}")
```

> ****Atención****
>
> No obstante, nos conviene utilizar funciones predefinidas para evitar
> errores.

Vamos a utilizar la función `factorint(n)` de la librería scipy.

``` python
sp.factorint(348)
```
